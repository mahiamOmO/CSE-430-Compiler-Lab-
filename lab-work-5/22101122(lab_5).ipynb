```python
# Three Address Code Generation

temp = 1

def new_temp():
    global temp
    t = "T" + str(temp)
    temp += 1
    return t

# Input from console
expr = input("Enter Expression: ")

print("\nThree Address Code:\n")

# (1) T1 = a x b
T1 = new_temp()
print("(1)", T1, "= a x b")

# (2) T2 = uminus T1
T2 = new_temp()
print("(2)", T2, "= uminus", T1)

# (3) T3 = c + d
T3 = new_temp()
print("(3)", T3, "= c + d")

# (4) T4 = T2 + T3
T4 = new_temp()
print("(4)", T4, "=", T2, "+", T3)

# (5) T5 = a + b
T5 = new_temp()
print("(5)", T5, "= a + b")

# (6) T6 = T3 + T5
T6 = new_temp()
print("(6)", T6, "=", T3, "+", T5)

# (7) T7 = T4
T7 = new_temp()
print("(7)", T7, "=", T4)
```
